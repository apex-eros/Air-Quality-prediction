```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python

from sklearn.model_selection import cross_val_score
mse=cross_val_score(lr,x_train,y_train,scoring='neg_mean_squared_error',cv=10)
np.mean(mse)
```

```python
#PREDICTION
y_pred=lr.predict(x_train)
y_pred
```

```python
sb.displot(y_train-y_pred,kind='kde')
```

```python
plt.scatter(y_pred,y_train)
plt.plot(y_test,y_test)
plt.xlabel("ACTUAL")
plt.ylabel("PREDICTED")
plt.show()
```

```python
y_pred_test=lr.predict(x_test)
y_pred_test

mae=metrics.mean_absolute_error(y_test,y_pred_test)
mae
```

```python
# ON TRAINING DATASET
y_predict = lr.predict(x_train)
rmse = (np.sqrt(mean_squared_error(y_train,y_predict)))
print("The model performance for training dataset")
print('RMSE is {}'.format(rmse))
print("\n")
```

```python


# ON TESTING DATASET

y_predict = lr.predict(x_test)
rmse = (np.sqrt(mean_squared_error(y_test,y_predict)))
print("The model performance for testing dataset")
print('RMSE is {}'.format(rmse))
```

```python
import numpy as np

# Replace this with your actual ML model for AQI prediction
def predict_aqi_lr(lr, input_data):

    return lr.predict([input_data])[0]

def generate_random_gas_values():
    # Generate random values for gases
    pm25 = np.random.uniform(0, 50)
    pm10 = np.random.uniform(0, 50)
    co = np.random.uniform(0, 5)
    co2 = np.random.uniform(300, 1000)
    no2 = np.random.uniform(0, 20)
    so2 = np.random.uniform(0, 10)
    o3 = np.random.uniform(0, 100)

    return pm25, pm10, co, co2, no2, so2, o3

def user_input_and_predict_aqi():
    # Get user input
    location = int(input("Enter location: "))
    time = int(input("Enter time (24-hour format): "))

    # Generate random gas values
    pm25, pm10, co, co2, no2, so2, o3 = generate_random_gas_values()

    # Predict AQI using linear regression model
    input_data = [location,time,pm25, pm10, co, co2, no2, so2, o3]
    aqi_lr = predict_aqi_lr(lr, input_data)


    print(f"\nPredicted AQI using Linear Regression model: {aqi_lr}")

# Call the function to test
user_input_and_predict_aqi()
```
